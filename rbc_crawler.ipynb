```python
import json
import requests
import bs4

class RBCparser():
  def __init__(self, root_url='https://www.rbc.ru/'):

    self.resp = requests.get('https://www.rbc.ru/rbcfreenews/64aa633e9a7947a975b3edc2?from=newsfeed')
    self.soup = bs4.BeautifulSoup(self.resp.content, 'html.parser')

  def find_links(self, soup=None):

    if not soup:
      soup = self.soup

    page_links = list()
    for link in soup.find_all('a'):
        if link.get('href'):
            if 'https' in link.get('href') and 'www.rbc.ru' in link.get('href'):
                page_links.append(link.get('href'))
    return page_links

  def find_all_links(self):

    all_links_final = list()

    for link in self.find_links():
        resp = requests.get(link)
        soup = bs4.BeautifulSoup(resp.content, 'html.parser')
        extr_l = self.find_links(soup)
        all_links_final.extend(extr_l)

    return all_links_final

  def find_text(self, links):
    res = {}
    for link in links:
      resp = requests.get(link)
      soup = bs4.BeautifulSoup(resp.content, 'html.parser')
      res[link] = ""
      for text in soup.find_all('p'):
        res[link] += " " + text.get_text(strip=True)
    return res


  def find_title(self, links):
    res = {}
    for link in links:
      resp = requests.get(link)
      soup = bs4.BeautifulSoup(resp.content, 'html.parser')
      for text in soup.find_all('title'):
        res[link] = text.get_text(strip=True)
    return res


  def find_images(self, links):
    res = {}
    for link in links:
      resp = requests.get(link)
      soup = bs4.BeautifulSoup(resp.content, 'html.parser')
      res[link] = []
      for img in soup.find_all('img'):
        res[link].append(img.get('src'))
    return res


  def save_dict(self, dct, file_name):
    open(file_name, "w").write(json.dumps(dct))
```

```python
rbc = RBCparser()
```

```python
links = rbc.find_links()
rbc.save_dict(links, 'links.txt')
```

```python
all_links = rbc.find_all_links()
rbc.save_dict(all_links, 'all_links.txt')
```

```python
texts = rbc.find_text(links)
rbc.save_dict(texts, 'texts.txt')
```
